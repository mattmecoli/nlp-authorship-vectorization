## Preprocessing

NOTE: THE FOLLOWING CODE IS NOT YET OPTIMIZED TO PULL OUT PARAGRAPHS

```python
# Cleaning and Tokenizing Words for train texts

train_folder = Path("rawtextfiles")

i = 0

clean_books = []

for filename in train_folder.iterdir():
    with open(str(filename), encoding='utf-8', errors='ignore') as fhand:
        text = fhand.readlines()
        clean_text = []
        for line in text:
            clean_line = re.sub('[^A-Za-z0-9 ]+', '', line)
            words = clean_line.split(" ")
            for word in words[:]:
                if len(word) != 0:
                    clean_text.append(word.lower())
        clean_books.append(clean_text[300:])
#         clean_books.append(clean_text)
        
        
for i in clean_books:
    if len(i) < 1:
        clean_books.remove(i)        

print(clean_books[0][:50])
print(clean_books[5][:50])
print(clean_books[10][:50])
```
